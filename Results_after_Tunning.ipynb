```
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns

bar_Width = 0.2
plt.figure(figsize=(15,10))
#data
bars1 = [0.95,0.98]
bars2 = [0.81,0.9]
bars3 = [0.44,0.73]
x= np.arange(len(bars1))

# Make the plot
sns.plt.bar(x,  bars1, color='steelblue', width=bar_Width, zorder=2)
sns.plt.bar(x + bar_Width,bars2, color='lightcoral',width=bar_Width,zorder=2)
sns.plt.bar(x + bar_Width*2,bars3, color='darkseagreen', width=bar_Width, zorder=2)

#labels
plt.xticks(x+ bar_Width,["Before","After"])
plt.title("Before vs After Tunning",fontsize=25)
plt.xlabel("Before and After for year1",fontsize=15)
plt.ylabel("Accuracy,Recall and Precision",fontsize=15)

#legend
patch1 = mpatches.Patch(color="steelblue",label="Accuracy")
patch2 = mpatches.Patch(color="lightcoral",label="Recall")
patch3 = mpatches.Patch(color="darkseagreen",label="Precision")
plt.legend(handles=[patch1,patch2,patch3])
plt.grid(axis='y')
```
