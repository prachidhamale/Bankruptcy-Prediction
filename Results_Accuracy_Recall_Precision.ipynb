```
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

bar_Width = 0.15
plt.figure(figsize=(20,10))
#data
bars1_accuracy = [0.95,0.99,.98,.68,0.63]
bars2_recall = [0.83,0.89,.88,.62,0.52]
bars3_precision= [0.47,0.86,.78,.20,0.45]
x= np.arange(len(bars1_accuracy ))

# Make the plot
plt.bar(x,  bars1_accuracy , color='green', width=bar_Width, zorder=2,label="Accuracy")
plt.bar(x + bar_Width,bars2_recall, color='red',width=bar_Width,zorder=2,label="recall")
plt.bar(x + bar_Width*2,bars3_precision, color='orange', width=bar_Width, zorder=2,label="precision")

#labels
plt.xticks(x+ bar_Width,["Mean-xgboost","Mean-RF","Mean-DT","Mean-svm","Mean-LR"])
plt.title("For Year 1")

#legend
green_patch = mpatches.Patch(color="green",label="Accuracy")
red_patch = mpatches.Patch(color="red",label="Recall")
orange_patch = mpatches.Patch(color="orange",label="Precision")
plt.legend(handles=[green_patch,red_patch,orange_patch])
```
